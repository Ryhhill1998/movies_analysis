```python
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
load_dotenv()
movies_csv_file_path = os.getenv("MOVIES_CSV_FILE_PATH")
```

```python
movies = pd.read_csv(movies_csv_file_path)
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
# TASK 1: Find highest rated movie genre.
# Movies must have at least 100 votes
```

```python
def get_top_genres_by_column(movies: pd.DataFrame, column: str, count: int = 5):
    movies_grouped_by_genre = movies.groupby("genre")
    average_column_value = movies_grouped_by_genre[[column]].mean()
    top_genres = average_column_value.nlargest(count, columns=[column])
    return top_genres
```

```python
def plot_bar_chart(x, height, title: str, xlabel: str, ylabel: str):
    fig, ax = plt.subplots()
    
    ax.bar(x=x, height=height)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_ylim(bottom=height.min() * 0.95, top=height.max() * 1.025)
    
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Genres are comma-separated strings - splitting so that each movie appears once for each of its genres
movies_by_genre = movies.copy()
movies_by_genre["genres"] = movies_by_genre["genres"].str.split(", ")
movies_by_genre = movies_by_genre.explode("genres").rename(columns={"genres": "genre"}).reset_index(drop=True)
movies_by_genre = movies_by_genre.drop_duplicates(subset=["id", "genre"])
```

```python
movies_with_min_votes = movies_by_genre[movies_by_genre["vote_count"] >= 100]
top_genres_by_vote_average = get_top_genres_by_column(movies=movies_with_min_votes, column="vote_average", count=5)
top_genres_by_vote_average
```

```python
plot_bar_chart(
    x=top_genres_by_vote_average.index, 
    height=top_genres_by_vote_average["vote_average"], 
    title="Top 5 Genres by Movie Rating", 
    xlabel="Genre", 
    ylabel="Average movie rating"
)
```

```python
top_genres_by_revenue = get_top_genres_by_column(movies=movies_by_genre, column="revenue", count=5)
top_genres_by_revenue
```

```python
plot_bar_chart(
    x=top_genres_by_revenue.index, 
    height=top_genres_by_revenue["revenue"], 
    title="Top 5 Genres by Movie Revenue", 
    xlabel="Genre", 
    ylabel="Average movie revenue"
)
```

```python
top_genres_by_budget = get_top_genres_by_column(movies=movies_by_genre, column="budget", count=5)
top_genres_by_budget
```

```python
plot_bar_chart(
    x=top_genres_by_budget.index, 
    height=top_genres_by_budget["budget"], 
    title="Top 5 Genres by Movie Budget", 
    xlabel="Genre", 
    ylabel="Average movie budget"
)
```
