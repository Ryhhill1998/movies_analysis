```python
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
load_dotenv()
movies_csv_file_path = os.getenv("MOVIES_CSV_FILE_PATH")
```

```python
movies = pd.read_csv(movies_csv_file_path)
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
# TASK 1: Find highest rated movie genre.
# Movies must have at least 100 votes
```

```python
# Get all movies with min 100 votes
movies_with_min_votes = movies[movies["vote_count"] >= 100]
```

```python
# Genres are comma-separated strings - splitting so that each movie appears once for each of its genres
movies_with_min_votes["genres"] = movies_with_min_votes["genres"].str.split(", ")
movies_with_min_votes = movies_with_min_votes.explode("genres").rename(columns={"genres": "genre"})
movies_with_min_votes = movies_with_min_votes.drop_duplicates(subset=["id", "genre"])
movies_with_min_votes.head()
```

```python
# group movies by genre
movies_grouped_by_genre = movies_with_min_votes.groupby("genre")

# get average rating for each genre
average_rating_by_genre = movies_grouped_by_genre[["vote_average"]].mean().sort_values(by="vote_average")
average_rating_by_genre
```

```python
# add average rating for each genre to bar chart
fig, ax = plt.subplots()

fig.set_size_inches(15, 8)

ax.bar(x=average_rating_by_genre.index, height=average_rating_by_genre["vote_average"])
ax.set_xlabel("Genre")
ax.set_ylabel("Average movie rating")
ax.set_ylim(bottom=5.5)

plt.xticks(rotation=90)

plt.show()
```
